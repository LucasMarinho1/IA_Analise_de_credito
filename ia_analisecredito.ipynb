```python
import pandas as pd
arquivo = pd.read_csv('baseDb/DadosAnaliseDeCredito.csv')
arquivo.head()
```

```python
print(arquivo.isnull().sum())

dados_duplicados = arquivo.duplicated().sum()
print(f'\nTemos {dados_duplicados} dados duplicados em nosso dataframe')
```

```python
arquivo.shape
```

```python
arquivo = arquivo.drop_duplicates()
arquivo.duplicated().sum()
arquivo
```

```python
metadados = pd.DataFrame({'nome': arquivo.columns, 'tipo': arquivo.dtypes})
metadados
```

```python
qtd_categorias = []

# faz um for ao longo dos nomes das variáveis
for nome in metadados['nome']:
  n = arquivo[nome].nunique() # conta o número de categorias dessa variável
  qtd_categorias.append(n) # acumula essa informação na lista

# atribui a lista à variável
metadados['qtd_categorias'] = qtd_categorias

metadados['qtd_categorias']
```

```python
variaveis_apropriadas = pd.get_dummies(arquivo, columns=['sexo', 'posse_de_veiculo', 'posse_de_imovel',
       'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)

variaveis_apropriadas.head(5)
```

```python
start = 400
end = 405
variaveis_apropriadas['mau'] = variaveis_apropriadas['mau'].replace('False', 0)
variaveis_apropriadas['mau'] = variaveis_apropriadas['mau'].replace('True', 1)
y = variaveis_apropriadas['mau']
x = variaveis_apropriadas.drop('mau', axis = 1)
variaveis_apropriadas.head()
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
print(variaveis_apropriadas.shape, x_treino.shape, x_teste.shape, y_treino.shape, y_teste.shape)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
resultado = modelo.score(x_teste, y_teste)
print("Acurácia:", resultado)
```
